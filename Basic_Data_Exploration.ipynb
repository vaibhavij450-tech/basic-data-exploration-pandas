```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\HP\Downloads\archive\tshirts.csv")

# Explore data
print(df.head())

print(df.tail())

print("Shape:", df.shape)

print("Columns:")
print(df.columns)

print("Data types:")
print(df.dtypes)

# Missing values
print("Missing values:")
print(df.isnull().sum())

# Remove duplicates
df = df.drop_duplicates()

# Create derived column
df["popularity_score"] = df["rating"] * df["ratings_count"]

# Save cleaned dataset
df.to_csv("cleaned_tshirts.csv", index=False)

print("Assignment completed successfully")
```
